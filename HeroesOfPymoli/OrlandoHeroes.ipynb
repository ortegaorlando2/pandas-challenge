```python
Gender_df=Player_df.groupby(by='Gender').count().reset_index()
Count=Gender_df['SN']
Percentage=Gender_df['SN']*100/numrows
```

```python
Gender_df['Count']=Count.values
Gender_df['Percentage']=Percentage.values
Gender_df[['Gender','Count','Percentage']].style.hide_index()
```

```python
PlayerStats_df= pd.DataFrame({'Age':Player_df['Age'].describe()})
PlayerStats_df=PlayerStats_df.reset_index()
PlayerStats_df
```

```python
Bins=pd.IntervalIndex.from_tuples([(0,9),(9,14),(14,19),(19,24),(24,29),(29,34),(34,39),(39,45)])
purchased_data['Age Bins'] = pd.cut(purchased_data['Age'], Bins)
```

```python
BinAge_df=purchased_data.groupby(['Age Bins']).count().reset_index()
AgeCount=BinAge_df['Age']
BinAge_df['Count']=AgeCount.values
```

```python
PriceBin=purchased_data.groupby('Age Bins').mean()
BinAge_df['Avg Price']=PriceBin['Price'].values.round(2)
```

```python
RevenueBin=purchased_data.groupby('Age Bins').sum()
BinAge_df['Revenue']=RevenueBin['Price'].values.round(2)
```

```python
Player_df['Age Bins'] = pd.cut(Player_df['Age'], Bins)
```

```python
Countperson=Player_df.groupby(['Age Bins']).count()
BinAge_df['Persons']=Countperson['Price'].values.round(2)
```

```python
BinAge_df['$/Person']=BinAge_df['Revenue'].div(BinAge_df['Persons'],axis=0).round(2)
```

```python
BinAge_df[{'Age Bins':['Age Bins'],'Count':['Count'],'Avg Price':['Avg Price'],'Revenue':['Revenue'],'$/Person':['$/Person']}].style.hide_index()
```
